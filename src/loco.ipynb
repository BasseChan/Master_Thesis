```python
from pathlib import Path

import soundfile as sf
from espnet.espnet2.bin.enh_inference import SeparateSpeech


separation_model = SeparateSpeech(
    train_config='espnet/egs2/wsj0_2mix/enh1/exp/enh_train_enh_tflocoformer_pretrained/config.yaml',
    model_file='espnet/egs2/wsj0_2mix/enh1/exp/enh_train_enh_tflocoformer_pretrained/valid.loss.ave_5best.pth',
    normalize_output_wav=True,
    device="cuda:0",
)
```

```python
import librosa

mix, sample_rate = sf.read('../data/2mix_val/mixed/0.wav')
mix = librosa.resample(mix, orig_sr=sample_rate, target_sr=8000)
mix /= mix.std(axis=-1)

spk1, spk2 = separation_model(mix[None], 8000)  # list of numpy arrays
```

```python
import IPython.display as ipd

# Odtwarzanie spk1
ipd.display(ipd.Audio(spk1.squeeze(), rate=8000))

# Odtwarzanie spk2
ipd.display(ipd.Audio(spk2.squeeze(), rate=8000))
```
